```python
#Splitting the data into independent (X) and dependent (Y) data sets.
from sklearn.model_selection import train_test_split
X = cancer.data
Y = cancer.target
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25)
```

We know that there are uneven number of occurances of benign and malign state in the data, as such our model might 
train more on the  state that occurs more frequently 

```python
# Check the distribution of data
print(Y.mean(), Y_test.mean(), Y_train.mean())
```

```python
#There is little difference but lets try to remove it
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, stratify = Y, random_state = 0)
print(Y.mean(), Y_test.mean(), Y_train.mean())
# Now the distribution of both the states are equal
```

```python
#Importing the loistic regression
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()
lg.fit(X_train, Y_train)
```

```python
# Training model
from sklearn.metrics import accuracy_score
pred_train = lg.predict(X_train)
accuracy_train = accuracy_score(Y_train, pred_train) 
print("Accuracy is : ", accuracy_train*100)
```

```python
#Testing model
pred_test = lg.predict(X_test)
accuracy_test = accuracy_score(Y_test, pred_test) 
print("Accuracy is : ", accuracy_test*100)
```

```python
# Lets test for different models
def models(X_train, Y_train):

  #Decision Tree
  from sklearn.tree import DecisionTreeClassifier
  tree = DecisionTreeClassifier(criterion = "entropy", random_state = 0)
  tree.fit(X_train, Y_train)

  #Random Forest Classifier
  from sklearn.ensemble import RandomForestClassifier
  forest = RandomForestClassifier(n_estimators = 10, criterion = "entropy", random_state = 0)
  forest.fit(X_train, Y_train)

  #Printing accuracy of above two training models
  print("[0] Accuracy of Decision Tree in training is : ", tree.score(X_train, Y_train))
  print("[1] Accuracy of Random Forest in training is : ", forest.score(X_train, Y_train))
  
  return tree, forest
```

```python
model = models(X_train, Y_train)
#We are getting excellent predictions on Decision Tree and Random Forest Classifier on training data
```

```python
# getting predictions for 
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
  print("Model", i)
  print(classification_report(Y_test, model[i].predict(X_test)))
  print(accuracy_score(Y_test, model[i].predict(X_test)))
  print()
```

Here we can clearly see that on training data decision tree and Random Forest does far better predictions on the training data than Logistic Regression. However, for test data Random Forest and Logistic Regression were able to generalize the data properly hence it does far better than Decision Tree.
